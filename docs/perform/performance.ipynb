```python
import matplotlib.pyplot as plt
import pandas as pd

lm = pd.read_csv('./Lm_res.csv', header=None)
logi = pd.read_csv('./Logistic_res.csv', header=None)

x = [1, 2]
label = ['abess','sklearn']
color = ['#00AF91', '#FFCC29']
```

```python
lm_time = plt.bar(x = x, height = lm.iloc[4, 0:2], tick_label = label, color = color)
for rect in lm_time:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2.0, height - 0.05, height, ha='center')

plt.ylabel('time(s)')
plt.title('lm_time')
plt.savefig('./lm_time.png')
```

```python
logi_time = plt.bar(x = x, height = logi.iloc[4, 0:2], tick_label = label, color = color)
for rect in logi_time:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2.0, height - 0.1, height, ha='center')

plt.ylabel('time(s)')
plt.title('logi_time')
plt.savefig('./logi_time.png')
```
